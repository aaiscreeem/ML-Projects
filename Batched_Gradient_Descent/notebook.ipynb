```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


df_train, df_cv = train_test_split(df, test_size=0.25, random_state=42)

# Helper function to compute necessary sums for training or CV set
def compute_sums(df):
    sumx = df['X'].sum()
    sumy = df['y'].sum()
    sumx2 = (df['X'] * df['X']).sum()
    sumxy = (df['X'] * df['y']).sum()
    sumy2 = (df['y'] * df['y']).sum()
    return sumx, sumy, sumx2, sumxy, sumy2

# Compute sums for the training set
n_train = df_train.shape[0]
sumx_train, sumy_train, sumx2_train, sumxy_train, sumy2_train = compute_sums(df_train)

# Compute sums for the cross-validation set
n_cv = df_cv.shape[0]
sumx_cv, sumy_cv, sumx2_cv, sumxy_cv, sumy2_cv = compute_sums(df_cv)

# Define the cost function J for both training and CV sets
def J(m, c, n, sumx, sumy, sumx2, sumxy, sumy2):
    return (m*m*sumx2 + c*c*n + 2*c*m*sumx + sumy2 - 2*m*sumxy - 2*c*sumy) / (2*n)

# Gradient Descent with Training and CV Losses
def gradient_descent(w0, w1, learning_rate, tol, max_epochs):
    m = w0
    c = w1
    J_values_train = []
    J_values_cv = []

    J_train_prev = 0
    J_train_curr = J(m, c, n_train, sumx_train, sumy_train, sumx2_train, sumxy_train, sumy2_train)
    J_values_train.append(J_train_curr)

    J_cv_curr = J(m, c, n_cv, sumx_cv, sumy_cv, sumx2_cv, sumxy_cv, sumy2_cv)
    J_values_cv.append(J_cv_curr)

    for i in range(max_epochs):
        temp_m = m - learning_rate * (m * sumx2_train + c * sumx_train - sumxy_train) / n_train
        c -= learning_rate * (m * sumx_train + n_train * c - sumy_train) / n_train
        m = temp_m

        J_train_prev = J_train_curr
        J_train_curr = J(m, c, n_train, sumx_train, sumy_train, sumx2_train, sumxy_train, sumy2_train)
        J_values_train.append(J_train_curr)

        J_cv_curr = J(m, c, n_cv, sumx_cv, sumy_cv, sumx2_cv, sumxy_cv, sumy2_cv)
        J_values_cv.append(J_cv_curr)

        if abs(J_train_curr - J_train_prev) < tol:
            print("GD - Learning Rate:", learning_rate, "Epochs:", i)
            break

    plt.plot(J_values_train, label='Training Loss')
    plt.plot(J_values_cv, label='Validation Loss')
    plt.xlabel("Iteration")
    plt.ylabel("Cost Function (J)")
    plt.title(f"Gradient Descent - Learning Rate {learning_rate}")
    plt.legend()
    plt.show()

    return m, c, J_train_curr, J_cv_curr

# Stochastic Gradient Descent with Training and CV Losses
def SGD(w0, w1, learning_rate, tol, max_epochs, batch_size, k, decay_rate):
    m = w0
    c = w1
    J_values_train = []
    J_values_cv = []

    arr = []

    for i in range(0, k * batch_size, batch_size):
        df1 = df_train.iloc[i:i + batch_size]
        n1 = df1.shape[0]
        sumx, sumy, sumx2, sumxy, sumy2 = compute_sums(df1)

        J_last = J(m, c, n1, sumx, sumy, sumx2, sumxy, sumy2)
        temp_m = m - learning_rate * (m * sumx2 + c * sumx - sumxy) / n1
        c -= learning_rate * (m * sumx + n1 * c - sumy) / n1
        m = temp_m
        J1 = J(m, c, n1, sumx, sumy, sumx2, sumxy, sumy2)

        arr.append(abs(J1 - J_last))
        J_values_train.append(J1)

        J_cv_curr = J(m, c, n_cv, sumx_cv, sumy_cv, sumx2_cv, sumxy_cv, sumy2_cv)
        J_values_cv.append(J_cv_curr)

    num = sum(arr)
    flag = 0
    for epoch in range(max_epochs):
        learning_rate /= (1 + decay_rate * epoch)
        df2 = df_train.sample(frac=1).reset_index(drop=True)
        for i in range(0, n_train, batch_size):
            df1 = df2.iloc[i:i + batch_size]
            if num / k < tol:
                print(f"SGD - Learning Rate: {learning_rate}, Batch Size: {batch_size}, Epochs: {epoch}")
                flag = 1
                break

            n1 = df1.shape[0]
            sumx, sumy, sumx2, sumxy, sumy2 = compute_sums(df1)
            J_last = J(m, c, n1, sumx, sumy, sumx2, sumxy, sumy2)
            temp_m = m - learning_rate * (m * sumx2 + c * sumx - sumxy) / n1
            c -= learning_rate * (m * sumx + n1 * c - sumy) / n1
            m = temp_m
            J1 = J(m, c, n1, sumx, sumy, sumx2, sumxy, sumy2)

            arr.append(abs(J1 - J_last))
            num += arr[-1] - arr[0]
            arr.pop(0)
            J_values_train.append(J1)

            J_cv_curr = J(m, c, n_cv, sumx_cv, sumy_cv, sumx2_cv, sumxy_cv, sumy2_cv)
            J_values_cv.append(J_cv_curr)

        if flag == 1:
            break

    plt.plot(J_values_train, label='Training Loss')
    plt.plot(J_values_cv, label='Validation Loss')
    plt.xlabel("Iteration")
    plt.ylabel("Cost Function (J)")
    plt.title(f"SGD - Learning Rate {learning_rate}, Batch Size {batch_size}")
    plt.legend()
    plt.show()

    return m, c, J_values_train[-1], J_values_cv[-1]

# Initialize parameters
w0 = 1  # slope
w1 = 1  # constant
tol = 0.00001
max_epochs = 100000
k = 24  # final after experimentation
decay_rate = 0.1

# Define learning rates and batch sizes
learning_rates = [0.1, 0.01, 0.001]
batch_sizes = [1, 10, 100, 1000]

# Gradient Descent Experimentation
for lr in learning_rates:
    m, c, J_train_final, J_cv_final = gradient_descent(w0, w1, lr, tol, max_epochs)
    print(f"GD - Learning Rate: {lr}, Slope (m): {m}, Constant (c): {c}, Training Cost: {J_train_final}, Validation Cost: {J_cv_final}")

# Stochastic Gradient Descent Experimentation
for lr in learning_rates:
    for bs in batch_sizes:
      if bs!=1000:
        m, c, J_train_final, J_cv_final = SGD(w0, w1, lr, tol, max_epochs, bs, k, decay_rate)
      else:
        m, c, J_train_final, J_cv_final = SGD(w0, w1, lr, tol, max_epochs, bs, 5, decay_rate)
      print(f"SGD - Learning Rate: {lr}, Batch Size: {bs}, Slope (m): {m}, Constant (c): {c}, Training Cost: {J_train_final}, Validation Cost: {J_cv_final}")
```
